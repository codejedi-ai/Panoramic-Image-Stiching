# NumPy

````python
# Double all elements
a1 = list(range(1080*1920*3))
a2 = np.arange(1080*1920*3)
````

````python
%%time
for i in range(1080*1920*3):
    a1[i] *= 2
````

````python
%%time
a2 *= 2
````

````python
# np.array or np.ndarray
a2 = a2.reshape(1080, 1920, 3)
print('ndim', a2.ndim)
print('shape', a2.shape)
print('dtype', a2.dtype)
````

## Types

````python
# Types
# # https://numpy.org/doc/stable/user/basics.types.html 

# Implicit type cast
a2[0,0,0] = 1.12
a2[0,0,0]
````

````python
# Explicit type cast
a2 = a2.astype(np.float32)
````

## Array functions

````python
# Creation
# np.zeros()
# np.ones()
# np.full()
````

````python
# Handy array functions
# https://numpy.org/doc/stable/reference/arrays.ndarray.html

# transpose
print('a2 shape', a2.shape)
print('Transpose', a2.T.shape)
# sort
a2.sort()
# stats
print('Min', a2.min())
print('Max', a2.max())
print('Mean', a2.mean())
print('Variance', a2.var())
print('Std', a2.std())
````

````python
# Axis specific operations
print('a2 shape', a2.shape)
a2.min(axis=-1)
print('a2 min shape', a2.min(axis=-1, keepdims=True).shape)
````

## Broadcast

1. Array vs. single value: element-wise operation
2. Array vs. array:
    1. Same size: element-wise
    2. Different number of dimensions: match from the right
        * Dimensions are equal
        * One of the dimensions is 1

````python
# Broadcast
# Array vs. value
# Operation is element-wise
# Normalize a2 to 0-1
# (a2 - min) / (max - min)
a2_norm = (a2 - a2.min()) / (a2.max() - a2.min())
print('Min', a2_norm.min())
print('Max', a2_norm.max())

plt.figure()
plt.imshow(a2_norm)
plt.show()
````

````python
# Array vs. array
a3 = np.random.rand(*a2.shape)
print('a2 shape', a2.shape)
print('a3 shape', a3.shape)
a23 = a2_norm + a3
a23 = (a23 - a23.min()) / (a23.max() - a23.min())
print('a2+a3 shape', a23.shape)

plt.figure(figsize=(8,3*6))
plt.subplot(311)
plt.imshow(a2_norm)
plt.subplot(312)
plt.imshow(a3)
plt.subplot(313)
plt.imshow(a23)
plt.show()
````

````python
a4 = np.random.rand(3)
print('a2 shape', a2.shape)
print('a4 shape', a4.shape)
a24 = a2_norm + a4
a24 = (a24 - a24.min()) / (a24.max() - a24.min())
print('a2+a4 shape', a24.shape)

plt.figure(figsize=(8,2*6))
plt.subplot(211)
plt.imshow(a2_norm)
plt.subplot(212)
plt.imshow(a24)
plt.show()
````

````python
a5 = np.random.rand(1920, 3)
print('a2 shape', a2.shape)
print('a5 shape', a5.shape)
a25 = a2_norm + a5
a25 = (a25 - a25.min()) / (a25.max() - a25.min())
print('a2+a5 shape', a25.shape)

plt.figure(figsize=(8,2*6))
plt.subplot(211)
plt.imshow(a2_norm)
plt.subplot(212)
plt.imshow(a25)
plt.show()
````

````python
a6 = np.random.rand(1080, 3)
print('a2 shape', a2.shape)
print('a6 shape', a6.shape)
print('a6+newaxis shape', a6[:,np.newaxis,:].shape)
a26 = a2_norm + a6[:,np.newaxis,:]
a26 = (a26 - a26.min()) / (a26.max() - a26.min())
print('a2+a6 shape', a26.shape)

plt.figure(figsize=(8,2*6))
plt.subplot(211)
plt.imshow(a2_norm)
plt.subplot(212)
plt.imshow(a26)
plt.show()
````

````python
# Excercise
a7 = np.random.rand(1920)
print('a2 shape', a2.shape)
print('a7 shape', a7.shape)
a27 = a2_norm + a7[:,np.newaxis]
a27 = (a27 - a27.min()) / (a27.max() - a27.min())
print('a2+a7 shape', a27.shape)

plt.figure(figsize=(8,2*6))
plt.subplot(211)
plt.imshow(a2_norm)
plt.subplot(212)
plt.imshow(a27)
plt.show()
````
